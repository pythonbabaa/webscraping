```python
import re
import urllib.request
import requests
from urllib.parse import urlsplit
from collections import deque
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.bahria.edu.pk/bulc/management-sciences-faculty/"
r = requests.get(url)
html_content = r.content
soup = BeautifulSoup(html_content, 'html.parser')
soup
```

```python
# Get the title of the HTML page
title = soup.title
print(type(title))  # 1. Tag
print(title)
print(title.string)
print(type(title.string))
print(type(soup))
```

```python
# Get all the anchor tags from the page

anchors = soup.find_all('a')
```

```python
all_links = set()

# Get all the links on the page

for link in anchors:
    if (link.get('href') != '#'):
        link = link.get('href')
        all_links.add(link)
        print(link)
```

```python
all_emails = set()

def scrape(url):
    
    r = requests.get(url)
    html_content = r.content
    html_content = html_content.decode()
    email_addresses = re.findall("[\w._%+-]{1,20}@[\w.-]{2,20}.[A-Za-z]{2,3}", html_content)
    for items in email_addresses:
        print(items)
        print(type(items))
        all_emails.add(items)
        
    return all_emails
```

```python
substring1 = "https://www.bahria.edu.pk/bulc"
substring2 = ".pdf"
substring3 = "students-resource"


for i in all_links:
    
    try:
        if (substring1 in i) & (substring2 not in i) & (substring3 not in i) & ("fee-structure" not in i):
            print("i is:   ", i)
            scrape(i)
    
    except:
        continue
```
